```python
#import necessary libraries

import pandas as pd
import numpy as np
import scipy.cluster.hierarchy as sch
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib as plt
```

```python
# Loading the dataset
data = pd.read_csv('environmental_socioeconomic.csv')
data.head()
```

```python
# Normalise the features using StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
print(pd.DataFrame(data_scaled, columns = data.columns).head())
```

```python
import matplotlib.pyplot as plt

# Make sure 'data_scaled' is defined before this line (for example, as a scaled version of your dataset)
plt.figure(figsize=(10, 6))
dendrogram = sch.dendrogram(sch.linkage(data_scaled, method='ward'))
plt.xlabel('Index')
plt.ylabel('Euclidean Distance')  # Corrected the spelling of 'Euclidean'
plt.title('Dendrogram of Hierarch')

```

```python
# Create th AgglomerativeClustering object with 7 clusters
hierarchical_clustering = AgglomerativeClustering(n_clusters=7, linkage='ward')
# Fit the model and assign the cluster labels to a new column 'cluster' in the data
data['cluster'] = hierarchical_clustering.fit_predict(data_scaled)
```

```python
print(data.head())
```

```python
sil_score = silhouette_score(data_scaled, data['cluster'])
print(f'Silhouette Score :{sil_score}')
```

```python
# Visualize using co2 emission and waste production
plt.figure(figsize=(8, 6))
sns.scatterplot(x= 'co2_emissions', y='waste_production', hue = 'cluster',
               data = data, palette = 'viridis', s=100, alpha= 0.7, edgecolor = 'k')
plt.title('Hierarchical clustering of environmental and socio economic factors')
plt.xlabel('co2 Emissions')
plt.ylabel('Waste production')
plt.legend(title = 'Cluster', bbox_to_anchor= (1.05, 1), loc = 'upper left')
plt.show()
```
